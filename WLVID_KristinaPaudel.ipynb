3.1 Problem 1A — Single Year HDI Exploration (Latest Year: 2022)

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv('/content/drive/MyDrive/AI Assessment/Human_Development_Index_Dataset.csv', encoding='latin1')

# Identify unique years
df['year'].unique()
```

```python
# Filter for year 2022
hdi_2022_df = df[df['year'] == 2022].copy()

# Save for further use
hdi_2022_df.to_csv("hdi_2022_df.csv", index=False)

hdi_2022_df.head(10)
```

```python
#Task2
hdi_2022_df.shape
hdi_2022_df.dtypes
```

```python
#Task3
# Check missing values
hdi_2022_df.isnull().sum()
# Remove duplicate rows
hdi_2022_df.drop_duplicates(inplace=True)

# Convert columns to numeric
numeric_cols = ['hdi', 'life_expectancy', 'gross_inc_percap']
for col in numeric_cols:
    hdi_2022_df[col] = pd.to_numeric(hdi_2022_df[col], errors='coerce')

# Handle missing values by dropping rows with missing HDI
hdi_2022_df = hdi_2022_df.dropna(subset=['hdi'])
```

```python
#Task4
hdi_2022_df['hdi'].mean()
hdi_2022_df['hdi'].median()
hdi_2022_df['hdi'].std()
# Highest and Lowest HDI
hdi_2022_df.loc[hdi_2022_df['hdi'].idxmax()][['country','hdi']]
hdi_2022_df.loc[hdi_2022_df['hdi'].idxmin()][['country','hdi']]
```

```python
#Task5
high_hdi = hdi_2022_df[hdi_2022_df['hdi'] > 0.8]

high_hdi_sorted = high_hdi.sort_values(by='gross_inc_percap', ascending=False)

high_hdi_sorted.head(10)
```

```python
#Task6
def categorize_hdi(hdi):
    if hdi < 0.55:
        return "Low"
    elif hdi < 0.7:
        return "Medium"
    elif hdi < 0.8:
        return "High"
    else:
        return "Very High"

hdi_2022_df['HDI_Category'] = hdi_2022_df['hdi'].apply(categorize_hdi)

hdi_2022_df[['country','hdi','HDI_Category']].head()
```

```python
# Save final CSV
save_path = "/content/drive/MyDrive/AI Assessment"
hdi_2022_df.to_csv(save_path + "/HDI_2022.csv", index=False)
```

3.2 Problem 1B — HDI Visualization and Trend Analysis (2020 – 2022)


```python
#Task1
hdi_2020_2022 = df[df['year'].isin([2020,2021,2022])].copy()

hdi_2020_2022.to_csv(save_path + "HDI_problem1B.csv", index=False)
hdi_2020_2022.head()
```

```python
#Task2
# Drop missing HDI, country, year
hdi_2020_2022.dropna(subset=['hdi','country','year'], inplace=True)

# Convert numeric
hdi_2020_2022['hdi'] = pd.to_numeric(hdi_2020_2022['hdi'], errors='coerce')
hdi_2020_2022['gross_inc_percap'] = pd.to_numeric(hdi_2020_2022['gross_inc_percap'], errors='coerce')

# Remove duplicates
hdi_2020_2022.drop_duplicates(inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

countries = ['Nepal','India','China','United States','Norway']
trend = hdi_2020_2022[hdi_2020_2022['country'].isin(countries)]

plt.figure(figsize=(8,5))
sns.lineplot(data=trend, x='year', y='hdi', hue='country', marker='o')
plt.title("HDI Trend (2020–2022)")
plt.xlabel("Year")
plt.ylabel("HDI")
plt.show()
```

```python
sa_list = ["Afghanistan", "Bangladesh", "Bhutan", "India", "Maldives", "Nepal", "Pakistan", "Sri Lanka"]
me_list = ["Bahrain", "Iran", "Iraq", "Israel", "Jordan", "Kuwait", "Lebanon", "Oman",
           "Palestine", "Qatar", "Saudi Arabia", "Syria", "United Arab Emirates", "Yemen"]

def assign_region(country):
    if country in sa_list:
        return 'South Asia'
    elif country in me_list:
        return 'Middle East'
    else:
        return 'Other'

hdi_2020_2022['region'] = hdi_2020_2022['country'].apply(assign_region)

region_avg = hdi_2020_2022.groupby(['region','year'])['hdi'].mean().reset_index()

plt.figure(figsize=(10,6))
sns.barplot(data=region_avg, x='region', y='hdi', hue='year')
plt.title("Average HDI by Region (2020–2022)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(data=hdi_2020_2022, x='year', y='hdi')
plt.title("HDI Distribution (2020–2022)")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.scatterplot(data=hdi_2020_2022, x='gross_inc_percap', y='hdi')
sns.regplot(data=hdi_2020_2022, x='gross_inc_percap', y='hdi', scatter=False, color='red')
plt.title("HDI vs GNI per Capita")
plt.show()
```

Problem 2: Advanced HDI Exploration


```python
# Task 1: Create South Asia Subset
south_asia = [
    "Afghanistan", "Bangladesh", "Bhutan", "India",
    "Maldives", "Nepal", "Pakistan", "Sri Lanka"
]
df_sa = df[df["country"].isin(south_asia)]
df_sa.to_csv("/content/drive/MyDrive/AI Assessment/HDI_SouthAsia.csv", index=False)

```

```python
#Task2
df_sa["Composite Score"] = (
    0.30 * df_sa["life_expectancy"] +
    0.30 * df_sa["gross_inc_percap"]
)

df_sa_sorted = df_sa.sort_values("Composite Score", ascending=False)
df_sa_sorted[["country", "Composite Score", "hdi"]]
top5 = df_sa_sorted.head(5)

plt.figure(figsize=(8,5))
plt.barh(top5["country"], top5["Composite Score"])
plt.xlabel("Composite Score")
plt.title("Top 5 South Asian Countries by Composite Score")
plt.gca().invert_yaxis()
plt.show()
```

```python
#Task3
def detect_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return (series < lower) | (series > upper)

df_sa["HDI_Outlier"] = detect_outliers(df_sa["hdi"])
df_sa["GNI_Outlier"] = detect_outliers(df_sa["gross_inc_percap"])
df_sa["Outlier"] = df_sa["HDI_Outlier"] | df_sa["GNI_Outlier"]
sns.scatterplot(
    data=df_sa,
    x="gross_inc_percap",
    y="hdi",
    hue="Outlier",
    palette={True: "red", False: "blue"}
)
plt.title("GNI per Capita vs HDI (Outliers Highlighted)")
plt.show()
```

```python
#Task4
from scipy.stats import pearsonr

corr_gender, _ = pearsonr(df_sa["gender_development"], df_sa["hdi"])
corr_life, _ = pearsonr(df_sa["life_expectancy"], df_sa["hdi"])

print("Gender Development vs HDI:", corr_gender)
print("Life Expectancy vs HDI:", corr_life)
sns.lmplot(data=df_sa, x="gender_development", y="hdi")
plt.title("Gender Development vs HDI")
plt.show()

sns.lmplot(data=df_sa, x="life_expectancy", y="hdi")
plt.title("Life Expectancy vs HDI")
plt.show()
```

```python
#Task5
df_sa["GNI_HDI_Gap"] = df_sa["gross_inc_percap"] - df_sa["hdi"]

df_sa.sort_values("GNI_HDI_Gap", ascending=False)[["country", "GNI_HDI_Gap"]]
top_pos = df_sa.sort_values("GNI_HDI_Gap", ascending=False).head(3)
top_neg = df_sa.sort_values("GNI_HDI_Gap").head(3)

plt.bar(top_pos["country"], top_pos["GNI_HDI_Gap"])
plt.title("Top 3 Positive GNI–HDI Gaps")
plt.show()

plt.bar(top_neg["country"], top_neg["GNI_HDI_Gap"])
plt.title("Top 3 Negative GNI–HDI Gaps")
plt.show()
```

Problem 3: Comparative Regional Analysis: South Asia vs Middle East

```python
#Task1
south_asia = [
    "Afghanistan", "Bangladesh", "Bhutan", "India",
    "Maldives", "Nepal", "Pakistan", "Sri Lanka"
]

middle_east = [
    "Bahrain", "Iran", "Iraq", "Israel", "Jordan", "Kuwait",
    "Lebanon", "Oman", "Palestine", "Qatar",
    "Saudi Arabia", "Syria", "United Arab Emirates", "Yemen"
]
df_2020_2022 = df[df["year"].isin([2020, 2021, 2022])]
df_sa = df_2020_2022[df_2020_2022["country"].isin(south_asia)]
df_me = df_2020_2022[df_2020_2022["country"].isin(middle_east)]
df_sa.to_csv("/content/drive/MyDrive/AI Assessment/HDI_SouthAsia_2020_2022.csv", index=False)
df_me.to_csv("/content/drive/MyDrive/AI Assessment/HDI_MiddleEast_2020_2022.csv", index=False)
```

```python
#Task2
sa_mean, sa_std = df_sa["hdi"].mean(), df_sa["hdi"].std()
me_mean, me_std = df_me["hdi"].mean(), df_me["hdi"].std()

print("South Asia - Mean:", sa_mean, "Std:", sa_std)
print("Middle East - Mean:", me_mean, "Std:", me_std)
```

```python
#Task3
sa_avg = df_sa.groupby("country")["hdi"].mean().sort_values()
me_avg = df_me.groupby("country")["hdi"].mean().sort_values()
sa_bottom3, sa_top3 = sa_avg.head(3), sa_avg.tail(3)
me_bottom3, me_top3 = me_avg.head(3), me_avg.tail(3)
plt.figure(figsize=(10,5))
plt.bar(sa_top3.index, sa_top3.values, label="South Asia Top", alpha=0.7)
plt.bar(sa_bottom3.index, sa_bottom3.values, label="South Asia Bottom", alpha=0.7)
plt.bar(me_top3.index, me_top3.values, label="Middle East Top", alpha=0.7)
plt.bar(me_bottom3.index, me_bottom3.values, label="Middle East Bottom", alpha=0.7)
plt.legend()
plt.title("Top & Bottom HDI Performers (2020–2022)")
plt.show()
```

```python
#Task4
metrics = ["gender_development", "life_expectancy", "gross_inc_percap"]

sa_means = df_sa[metrics].mean()
me_means = df_me[metrics].mean()

comparison = pd.DataFrame({
    "South Asia": sa_means,
    "Middle East": me_means
})

comparison.plot(kind="bar", figsize=(8,5))
plt.title("Metric Comparison: South Asia vs Middle East")
plt.ylabel("Average Value")
plt.show()
```

```python
#Task5
sa_range = df_sa["hdi"].max() - df_sa["hdi"].min()
me_range = df_me["hdi"].max() - df_me["hdi"].min()

sa_cv = sa_std / sa_mean
me_cv = me_std / me_mean

print("South Asia Range:", sa_range, "CV:", sa_cv)
print("Middle East Range:", me_range, "CV:", me_cv)
```

```python
#Task6
for region, data in [("South Asia", df_sa), ("Middle East", df_me)]:
    print(region)
    print("HDI vs Gender Development:", pearsonr(data["gender_development"], data["hdi"])[0])
    print("HDI vs Life Expectancy:", pearsonr(data["life_expectancy"], data["hdi"])[0])
sns.lmplot(data=df_sa, x="gender_development", y="hdi")
plt.title("South Asia: Gender Development vs HDI")
plt.show()

sns.lmplot(data=df_me, x="life_expectancy", y="hdi")
plt.title("Middle East: Life Expectancy vs HDI")
plt.show()
```
